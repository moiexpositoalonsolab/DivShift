```python
import dask.dataframe as dd
import pandas as pd
```

```python
states = ['alaska', 'arizona', 'baja_california', 'baja_california_sur', 'british_columbia', 
                  'california', 'nevada', 'oregon', 'sonora', 'washington', 'yukon']
ddf = dd.concat([dd.read_csv(f'../west_coast_dataset/{state_name}/observations_postGL.csv') for state_name in states])
ddf['year'] = dd.to_datetime(ddf['date']).dt.year
ddf['year'].compute()
```

```python
train_df = ddf[ddf['supervised'] & 
               (ddf['download_success'] == 'yes') & 
               ((ddf['year'] == 2019) | (ddf['year'] == 2020) | (ddf['year'] == 2021))].compute()
train_df
```
